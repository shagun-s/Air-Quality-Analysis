```python
df[['aqi_index','temp_c','humidity','windspeed_kph']].corr()
```

```python
def categorize_aqi(aqi):
    if aqi <= 50:
        return "Good"
    elif aqi <= 100:
        return "Satisfactory"
    elif aqi <= 200:
        return "Moderate"
    elif aqi <= 300:
        return "Poor"
    elif aqi <= 400:
        return "Very Poor"
    else:
        return "Severe"

df['AQI_Category'] = df['aqi_index'].apply(categorize_aqi)
```

```python
df['AQI_Category'].value_counts()
```

```python
df['Hour'] = df['datetime'].dt.hour
```

```python
df.groupby('Hour')['aqi_index'].mean()
```

```python
avg_aqi = df.groupby('location')['aqi_index'].mean().sort_values()

plt.figure(figsize=(12,6))
avg_aqi.plot(kind='bar')
plt.title("Average AQI by Location")
plt.show()
```

```python
monthly = df.groupby('Month_Name')['aqi_index'].mean()

monthly.plot(kind='line')
plt.title("Monthly AQI Trend")
plt.show()
```
